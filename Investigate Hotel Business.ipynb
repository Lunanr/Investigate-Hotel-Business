# Data Preprocessing

```python
#Check a null value from data
df.isnull().sum()
```

```python
#Fill a childern,agent,company data with 0 because this is intereger data type
#Fill a city data with "Unknown" because this is object data type
nan_value = {'childern' : 0,
             'city' : 'unknwon',
             'agent' : 0,
             'company' :0}
df_fill = df.fillna(nan_value)
```

```python
df_fill['guest'] = df_fill['adults']+df_fill['children']+df_fill['babies']
#Check total from guest there is a 0 values from the guest because it we must remove them
df_fill['guest'].value_counts()
```

```python
df_prep = df_fill[df_fill["guest"]>0].copy()
print('Before a preprocessing: ', df_fill.shape[0])
print('After a preprocessing: ', df_prep.shape[0])
```

# Analyst 1
## Analysis the hotel booking per month based by hotel type

```python
#Copy a pre-processing data result
df_da1 = df_prep.copy()

#Make a aggregation for number of hotel booking's per month
df_da1 = df_da1.groupby(['hotel','arrival_date_month'])['arrival_date_year'].agg(['nunique','count']).reset_index()
#Change a name for a column
df_da1.columns = ['hotel_type','arrival_date_month','unique_year','num_booking']
df_da1
```

look a data from october and september the data was take by 3 years ago, while other months take by 2 years ago, with that we must do normalization data 

```python
#Normalization data to get a average number of order in a month in a different years
df_da1['average_num_booking'] = round(df_da1['num_booking']/df_da1['unique_year'])
df_da1
```

month writing is better abbreviated to make visualization eaiser. Therefore, we make a new column the content is 3 letter from the name of month

```python
#Make a new variable for make a abbreviated month
month_order = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']
#Take a 3 letter from the name of month
df_da1['month'] = df_da1['arrival_date_month'].str[:3]
#Sort a month
df_da1['month'] = pd.Categorical(df_da1['month'],categories=month_order, ordered=True)

df_da1
```

```python
#set a figure size
plt.figure(figsize=(15,10))
#set a new lineplot
sns.lineplot(x='month',y='average_num_booking',hue='hotel_type',
             size='hotel_type',sizes=(3, 3), data=df_da1)

#Increase size x-y axis label
plt.tick_params(axis='both', which='major', labelsize=15)
plt.grid() #Make a gridline
#Set a title and font size on legend
plt.legend(title='Hotel Type', title_fontsize=15, prop={'size':13})

#set a title on a x/y axis
plt.xlabel('Arrival Month', fontsize=15)
plt.ylabel('Average Number of Booking', fontsize=15)
plt.ylim(0, 4800) #limit a y axis

#set a vertical line for highlighting insights
plt.axvline(4, ls='--', color='red')
plt.axvline(6, ls='--', color='red')
#set a text between vertical line
plt.text(x=4.5, y=4400, s='Holiday \nSeason', fontsize=16, color='red')
#set a block of color to the highlighted area
plt.stackplot(np.arange(4,7,1), [[4800]], color='indianred', alpha=0.3)

plt.axvline(9, ls='--', color='red')
plt.axvline(11, ls='--', color='red')
plt.text(x=9.5, y=4400, s='Holiday \nSeason', fontsize=16, color='red')
plt.stackplot(np.arange(9,12,1), [[4800]], color='indianred', alpha=0.3)

#set a title for a visualization data
plt.text(x=-0.5, y=5200, s='Hotel Bookings During Holiday Season',
         fontsize=20, fontweight='bold')
plt.text(x=-0.5, y=5000, s='City Hotel has a Increased Guest in May-jul and Oct-Dec',
         fontsize=18)

#set a layour for a visualization so as not to cut
plt.tight_layout()
#save a picture
plt.savefig('Investigate hotel business DA1.png', dpi=1000);
```

# Analyst 2

## Analysis of the effect of the duration of stay on the rate of cancellation of hotel bookings

```python
# Copy preprocessing data
df_da2 = df_prep.copy()
# Do aggregration to get a duration stay based on hotel type
df_da2["total_nights"] = df_da2["stays_in_weekend_nights"] + df_da2["stays_in_weekdays_nights"]
# Set a histplot to see a stay duration
sns.histplot(data=df_da2, x="total_nights", binwidth=5, color='red');
```

Based on the histogram above, it can be seen that the data distribution looks right-skewed. Therefore grouping is done to make the distribution more even. The grouping is done for a total number of nights which reach 3 weeks, so they are grouped from the 15th night to the 21st night (15-21). Next for total nights over 3 weeks (>21)

```python
# Data Grouping based on distribution data
df_da2['total_nights_group'] = np.where(df_da2['total_nights'] > 21, 21,
                                       np.where(df_da2['total_nights']>14, 15, df_da2['total_nights']))

# Make a agregration tabel
df_da2 = df_da2.groupby(['hotel','is_canceled','total_nights_group'])['arrival_date_month'].count().reset_index(name='num_bookings')
#Create a spreadsheet-style pivot table as a DataFrame
df_da2 = df_da2.pivot_table(index=['hotel','total_nights_group'], columns='is_canceled', values='num_bookings').reset_index()
df_da2['pct_canceled'] = round(df_da2[1]*100.0/(df_da2[0]+df_da2[1]),2)

df_da2.head(5)
```

```python
plt.figure(figsize=(15,10))
# Make plot data and a linear regression model fit
sns.regplot(x='total_nights_group', y='pct_canceled', data=df_da2[df_da2['hotel']=='City Hotel'], scatter=False, label='Trend City Tabel')
sns.regplot(x='total_nights_group', y='pct_canceled', data=df_da2[df_da2['hotel']=='Resort Hotel'], scatter=False, label='Trend City Tabel')
# Make barplot
sns.barplot(x='total_nights_group', y='pct_canceled', hue='hotel', data=df_da2)

plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(title='Hotel Type', title_fontsize=15, prop={'size':13})

# Make a customized-label on x-axis
plt.xticks(range(0, 17,1), labels=[str(x) for x in range(0,15,1)]+["15-21",">21"])
plt.xlabel('Total Nights Booked', fontsize=15)
plt.ylabel('Percentage of Booking Canceled (%)', fontsize=15)
plt.ylim(0, 100)
plt.xlim(0.5, 16.5)

plt.text(x=0.5, y=110, s="Positive Trend on Total Nights and Percentage of Booking Canceled",
        fontsize=20, fontweight='bold')
plt.text(x=0.5, y=102, s="The longer the total nights booked, the higher the percentage of booking being canceled, \nalso City Hotel has more steeper trend compared to Resort Hotel",
        fontsize=18);

plt.tight_layout()
plt.savefig('Investigate hotel businees DA2', dpi=1000);
```

# Analyst 3

## Analysis of the Influence of Booking Distance on Hotel Booking Cancellation Rates

```python
# Doing copy from preprocessing data
df_da3 = df_prep.copy()

sns.histplot(data=df_da3, x="lead_time", bins=50, color="red");
```

```python
df_da3['lead_times'] = np.where(df_da3['lead_time'] > 400, 401, df_da3['lead_time'])
bins = list(np.arange(0,max(df_da3['lead_times'])+30, 30))
labels = ['{0}-{1}'.format(str(bins[x]), str(bins[x+1])) for x in range(len(bins[:-1]))]
#Use cut when you need to segment and sort data values into bins
#Groupping lead_time for multiples in 30
df_da3['lead_time_group'] = pd.cut(df_da3['lead_times'], bins=bins, include_lowest=True, labels=labels)
```

```python
# Doing a aggregration data for count the amount of hotel bookings based on lead time group
df_da3 = df_da3.groupby(['hotel','lead_time_group', 'is_canceled'])['hotel'].count().reset_index(name='num_bookings')
# Create a spreadsheet-style pivot table as a DataFrame
df_da3 = df_da3.pivot_table(index=['hotel','lead_time_group'], columns='is_canceled', values='num_bookings').reset_index()
# Make a new columns
df_da3.columns = ['hotel_type','lead_time_group','not_canceled','canceled']
df_da3['all_bookings'] = df_da3['not_canceled'] + df_da3['canceled']
df_da3['pct_canceled'] = df_da3['canceled']*100.0/(df_da3['all_bookings'])

df_da3.head()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='lead_time_group', y='pct_canceled', hue='hotel_type', data=df_da3)

plt.tick_params(axis='both', which='major', labelsize=12)
plt.grid()
plt.legend(title='Hotel Type', title_fontsize=14, prop={'size':12}, loc=9)

plt.xlabel('Lead Time Groups (by 30 days)', fontsize=14)
plt.ylabel('Percentage of Booking Canceled (%)', fontsize=14)
plt.ylim(0, 100)
plt.xlim(-0.5, 13.5)

plt.axvline(0.5, ls='--', color='red')
plt.text(x=-0.2, y=80, s='1 Month', fontsize=18, color='red', rotation=90)
plt.stackplot(np.arange(-0.5,1.5,1), [[100]], color='indianred',alpha=0.3)

plt.axvline(8.5, ls='--', color='red')
plt.text(x=10, y=90, s='Around 1 Year', fontsize=18, color='red')
plt.stackplot(np.arange(8.5,14.5,1), [[100]], color='indianred', alpha=0.3)

plt.text(x=-0.5, y=108, s='Both Hotels Have the Lowest Canceled Ratio Within 1 Month of Leads Time',
         fontsize=20, fontweight='bold')
plt.text(x=-0.5, y=102, s='Resort Hotel has quite stagnant (in around 40%) canceled ratio, while City Hotel has quite high canceled \nratio (above 60%) with leads time around 1 year',
         fontsize=16)

plt.tight_layout()
plt.savefig('investigate_hotel_business_DA3.png', dpi=1000);
```
